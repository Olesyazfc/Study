```python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}
        self.avegare_grade = 0
        self.list_grades = []
        self.sum_grades = 0        
        
                         
    def add_courses(self, course_name):
        self.finished_course.append(course_name)   

    def av_grade(self):
        for courses_grades in list(self.grades.values()):
            for grade in courses_grades:
                self.sum_grades += grade
                self.list_grades.append(grade)
        self.avegare_grade = self.sum_grades / len(self.list_grades) 

    def __str__(self):
        self.av_grade()  
        student_inf = f'Имя: {self.name} \nФамилия: {self.surname} \nСредняя оценка за домашние задания: {self.avegare_grade :.1f} \nКурсы в процессе изучения: {",".join(self.courses_in_progress)} \nЗавершенные курсы: {",".join(self.finished_courses)}'
        return student_inf
               
    def __lt__(self, other):
        self.av_grade()
        other.av_grade()
        return self.avegare_grade > other.avegare_grade


class Mentor:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []
       
class Reviewer(Mentor):  
    def rate_hw(self, student, course, grade):
        if isinstance(student, Student) and course in self.courses_attached and course in student.courses_in_progress:
            if course in student.grades:
                student.grades[course] += [grade]
            else:
                student.grades[course] = [grade]
        else:
            return 'Ошибка'
    
    def __str__(self):
        reviwer_inf = f'Имя: {self.name} \nФамилия: {self.surname}'
        return reviwer_inf

class Lecturer(Mentor):       
    def __init__(self, name, surname):
        super().__init__(name, surname)
        self.grades = {}
        self.sum_grades = 0
        self.list_grades = []
        self.avegare_grade = 0

    def av_grades(self):
        for courses_grades in list(self.grades.values()):
            for grade in courses_grades:
                self.sum_grades += grade
                self.list_grades.append(grade)
            self.avegare_grade = self.sum_grades / len(self.list_grades) 

    def rate_hw(self, student, course, grade):
        if isinstance(student, Student) and course in self.courses_attached and student.courses_in_progress and 0 < grade < 11:
            if course in self.grades:
                self.grades[course] += [grade]
            else:
                self.grades[course] = [grade]
        else:
            return 'Ошибка'

    def __str__(self):
        self.av_grades()
        lecturer_inf = f'Имя: {self.name} \nФамилия: {self.surname} \nСредняя оценка за лекции: {self.avegare_grade}'
        return lecturer_inf

    def __lt__(self, other):
        self.av_grades()
        other.av_grades()
        return self.avegare_grade > other.avegare_grade


def av_all_students(students_list, course):
    av_students_list = []
    for student in students_list:
        if isinstance(student, Student) and course in student.finished_courses or course in student.courses_in_progress:
            student.av_grade()
            if student.avegare_grade != 0:
                av_students_list.append(student.avegare_grade)
    sum_st_list = sum(av_students_list) / len(av_students_list) 
    return round(sum_st_list, 1) 

def av_all_lectures(lectures_list, course):
    av_lectures_list = []
    for lecturer in lectures_list:
        if isinstance(lecturer, Lecturer) and course in lecturer.courses_attached:
            lecturer.av_grades()
            if lecturer.avegare_grade != 0:
                av_lectures_list.append(lecturer.avegare_grade)
    sum_lr_list = sum(av_lectures_list) / len(av_lectures_list) 
    return round(sum_lr_list, 1)


best_student = Student('Ruoy', 'Eman', 'your_gender')
best_student.courses_in_progress += ['Python']

worst_student = Student('Nice', 'Guy', 'some_gender')
worst_student.courses_in_progress += ['Python']
worst_student.courses_in_progress += ['Java']
 
cool_mentor = Mentor('Some', 'Buddy')
cool_mentor.courses_attached += ['Python']

cool_lecturer = Lecturer('Some', 'Buddy')
cool_lecturer.courses_attached += ['Python']

some_lecturer = Lecturer('This', 'Man')
some_lecturer.courses_attached += ['Python']

cool_reviewer = Reviewer('Some', 'Buddy')
cool_reviewer.courses_attached += ['Python']

some_reviewer = Reviewer('That', 'Man')
some_reviewer.courses_attached += ['Python']
some_reviewer.courses_attached += ['Java']

cool_reviewer.rate_hw(best_student, 'Python', 10)
cool_reviewer.rate_hw(best_student, 'Python', 10)
cool_reviewer.rate_hw(best_student, 'python', 10)

cool_lecturer.rate_hw(best_student, 'Python', 10)
cool_lecturer.rate_hw(best_student, 'Python', 10)
cool_lecturer.rate_hw(best_student, 'Python', 10)

some_reviewer.rate_hw(worst_student, 'Java', 2)
some_reviewer.rate_hw(worst_student, 'Java', 6)
some_reviewer.rate_hw(worst_student, 'Python', 3)

print(best_student.grades)
print(cool_lecturer.grades)

print(best_student)
print(cool_reviewer)
print(cool_lecturer)

print(cool_lecturer > some_lecturer)
print(worst_student > best_student)

students_list = [best_student, worst_student]
lectures_list = [some_lecturer, cool_lecturer]

print(av_all_students(students_list, 'Python'))
print(av_all_lectures(lectures_list, 'Python'))
```
